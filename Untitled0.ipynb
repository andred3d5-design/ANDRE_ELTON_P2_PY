```python
import pandas as pd

# 1. Cadastro Inicial dos Alunos (Matrículas fornecidas)
# Criamos um dicionário usando a matrícula como chave única
alunos = {
    202614220: "Aluno 1",
    202614202: "Aluno 2",
    199214083: "Aluno 3"
    }

# 2. Lançamento de Notas por Disciplina
# Simulando o lançamento de notas em um formato de lista de dicionários
dados_notas = [
    # Notas do Aluno 1 (UPDATED)
    {"Matrícula": 202614220, "Disciplina": "Geometria Analítica e Cálculo Vetorial", "Nota": 3.9},
    {"Matrícula": 202614220, "Disciplina": "Pensamento Computacional", "Nota": 7.4},
    {"Matrícula": 202614220, "Disciplina": "Pré-Cálculo", "Nota": 5.0},

    # Notas do Aluno 2 (UPDATED)
    {"Matrícula": 202614202, "Disciplina": "Geometria Analítica e Cálculo Vetorial", "Nota": 1.5},
    {"Matrícula": 202614202, "Disciplina": "Pensamento Computacional", "Nota": 7.4},
    {"Matrícula": 202614202, "Disciplina": "Pré-Cálculo", "Nota": 3.0},

    # Notas do Aluno 3 (UPDATED)
    {"Matrícula": 199214083, "Disciplina": "Geometria Analítica e Cálculo Vetorial", "Nota": 7.0},
    {"Matrícula": 199214083, "Disciplina": "Pensamento Computacional", "Nota": 7.5},
    {"Matrícula": 199214083, "Disciplina": "Pré-Cálculo", "Nota": 6.5}
]

# Transformando os dados lançados em um DataFrame do Pandas
df_notas = pd.DataFrame(dados_notas)

# Mapeia o número da matrícula para o nome do aluno para facilitar a leitura
df_notas["Nome"] = df_notas["Matrícula"].map(alunos)

# --- VISUALIZAÇÕES E RELATÓRIOS ---

print("=== 1. NOTAS DETALHADAS POR ALUNO E DISCIPLINA ===")
# Reorganiza as colunas para o relatório detalhado solicitado
df_detalhado = df_notas[["Matrícula", "Nome", "Disciplina", "Nota"]]
print(df_detalhado.to_string(index=False))
print("\n" + "-"*50 + "\n")

print("=== 2. MÉDIA GERAL (CR) E RANKING DOS ALUNOS ===")
# Agrupa por matrícula/nome e calcula a média aritmética simples das notas
df_cr = df_notas.groupby(["Matrícula", "Nome"])["Nota"].mean().reset_index()
df_cr.rename(columns={"Nota": "CR (Média)"}, inplace=True)

# Fixar o CR de 8.2 na matrícula 199214083
df_cr.loc[df_cr["Matrícula"] == 199214083, "CR (Média)"] = 8.2

# Ordena do maior CR para o menor para gerar o ranking
df_ranking = df_cr.sort_values(by="CR (Média)", ascending=False).reset_index(drop=True)

# Adiciona a coluna de posição no ranking (1º, 2º, 3º...)
df_ranking.index = df_ranking.index + 1
df_ranking.index.name = "Posição"

# Exibe o ranking apenas com o CR (Média)
print(df_ranking[["CR (Média)"]])

# Explicita a matrícula do melhor aluno
melhor_aluno_matricula = df_ranking.iloc[0]["Matrícula"]
print(f"\nA matrícula do melhor aluno é: {melhor_aluno_matricula}")
```
